# linear regresion model with one variable

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv(r"D:\Angel-Has-Fallen\Linear model\One-variable\custom_data.csv")
df
```

## we will use seaborn to visualise data on graph, using this you can easily visualise your data graphically 
**scattered plot**


```python
sns.lmplot(x="age", y="premium", data=df)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['age']], df['premium'])
```

# now we will predict on age 21 then lets see premium based on given age 

```python
reg.predict([[21]])
```

```python
reg.predict([[23]])
```

```python
reg.predict([[32]])
```

# lets see at 50 age 

```python
reg.predict([[50]])
```

#### this is slope which is also known as coefficient y =mx+c, here m is slope/coeffiecient

```python
reg.coef_
```

```python
reg.intercept_
```
